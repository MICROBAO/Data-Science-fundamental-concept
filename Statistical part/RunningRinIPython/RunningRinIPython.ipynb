# An IPython Notebook showing how to Run R in IPython

To run R within IPython,

(1)  get the rpy2 module for Python with a with something like the following Unix incantation:

     cd /Applications/anaconda    #  or wherever you installed anaconda
     easy_install rpy2   #  if you have a Mac, perhaps instead use  "sudo easy_install rpy2"

 (2) start an IPython notebook with something like the following an IPython incantation:

     %load_ext rmagic
     import rpy2 as Rpy

```python
%load_ext rmagic
import rpy2 as Rpy
```

That's it!  We now can run R within IPython.

```python
%R print( "Hello world" )     #  The '%R' annotation runs a single command in an R process
```

```python
%R hist(rnorm(10000, mean=100, sd=10), 100, col="cyan")
```

```python
%%R

hist(rnorm(10000, mean=100, sd=10), 100, col="cyan")    #  the '%%R' annotation runs an entire cell in an R process

mtext('we can add subheadings for the histogram')
```

```python
%%R 

not.installed <- function(pkg) !is.element(pkg, installed.packages()[,1])

if (not.installed("MASS"))  install.packages("MASS")  # we need the MASS package
```

```python
%%R
library(MASS)  #  load in the MASS package
print(library(help=MASS))   # print the help summary for what is in the package
```
