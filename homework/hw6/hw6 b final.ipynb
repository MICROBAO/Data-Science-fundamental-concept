```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import warnings
warnings.simplefilter('ignore', DeprecationWarning)
```

```python
data = pd.read_csv('/Users/Dongzhe/Desktop/cs249/hw6/data_science_job_train.csv')
data.shape
```

```python
data.count()
title=list(data.columns.values)
title=title[0:-1]
title
```

```python
data_rich = pd.concat([data.get(title),
                           pd.get_dummies(data.Class, prefix='Class')],axis=1)
data_rich.head()
```

```python
data_final =data_rich.drop('Class_nonDS', 1)

data_final.groupby('Class_DS').count()  
```

```python
target=data_final.Class_DS.values
type(target)
```

```python

features=data_final.get([   "d3js_tag" , "mysql_tag"  ,"api_tag", "data_modeling_tag",  "java_ee_tag","nosql_tag"  ,"frontend_tag"  ,"c_tag",'data','sql','hadoop','python','java','networking_tag','embedded_tag','machine_learning_tag','hacking_tag','javascript','linux','mongodb_tag','big_tag','amazon_web_services_tag','ruby','excel_tag','mysql','postgresql_tag'])

features
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(C=1)
```

```python
from sklearn.cross_validation import cross_val_score

scores = cross_val_score(logreg, features, target, cv=5, n_jobs=3,
                         scoring='accuracy')
print(scores.min(), scores.mean(), scores.max())
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.cross_validation import cross_val_score


gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.2,
                                subsample=.8, max_features=.25)
scores = cross_val_score(gb, features, target, cv=10, n_jobs=4,
                         scoring='accuracy')
print(scores.min(), scores.mean(), scores.max())
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)
scores = cross_val_score(rf,features, target, cv=5, n_jobs=4,
                         scoring='accuracy')
print(scores.min(), scores.mean(), scores.max())
```

```python
from sklearn.grid_search import GridSearchCV

gb = GradientBoostingClassifier(n_estimators=100, subsample=.8)

params = {
    
    'learning_rate':[0.1,0.05,0.25,0.2],
    'max_features':[0.2,0.4,0.6,0.8],
    'max_depth': [3, 4, 5],
}

gs = GridSearchCV(gb, params, cv=5, scoring='roc_auc', n_jobs=4)

gs.fit(features, target)

gs.cv
```
