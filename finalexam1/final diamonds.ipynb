```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt


import warnings
warnings.simplefilter('ignore', DeprecationWarning)

Diamonds = pd.DataFrame.from_csv('/Users/Dongzhe/Desktop/cs249/Final_Part01/Diamonds/Diamond_train.csv').reset_index()

Diamonds.head()
```

```python
Diamonds.count()
```

```python
rich_features = pd.concat([Diamonds.get(['x', 'y', 'z','Carat','Table','Depth']),
                           pd.get_dummies(Diamonds.Clarity, prefix='Clarity'),
                           pd.get_dummies(Diamonds.Color, prefix='Color'),
                           pd.get_dummies(Diamonds.Cut, prefix='Cut')],
                          axis=1)
target = Diamonds.Price.values
target.shape
```

```python
rich_features
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import cross_val_score

logreg = LogisticRegression(C=1)
scores = cross_val_score(logreg, rich_features, target, cv=5, scoring='accuracy')
print(scores.min(), scores.mean(), scores.max())
```

```python
from sklearn.cross_validation import train_test_split

features_train, features_test, target_train, target_test = train_test_split(
    rich_features, target, test_size=0.20, random_state=0)
```

```python
logreg.fit(features_train, target_train)
```

```python
target_predicted = logreg.predict(features_test)
 
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(target_test, target_predicted)
```

```python
logreg.coef_
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=150)
scores = cross_val_score(rf,rich_features, target, cv=5, n_jobs=4,
                         scoring='accuracy')
print(scores.min(), scores.mean(), scores.max())
```

```python
rich_features= rich_features.fillna(-1)
rich_features.head(6)
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test = train_test_split(rich_features.values)
```

```python
from sklearn.preprocessing import Imputer

imputer = Imputer(strategy='mean', missing_values=-1)

imputer.fit(X_train)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.cross_validation import cross_val_score

imputer = Imputer(strategy='mean', missing_values=-1)

classifier = GradientBoostingClassifier(n_estimators=100, learning_rate=0.5,
                                        subsample=.8, max_features=.5)

pipeline = Pipeline([
    ('imp', imputer),
    ('clf', classifier),
])
scores = cross_val_score(pipeline, rich_features, target, cv=5, n_jobs=4,
                         scoring='accuracy', )
print(scores.min(), scores.mean(), scores.max())
```

```python
from sklearn.grid_search import GridSearchCV

params = {
    'clf__learning_rate':[0.05,0.1,0.5],
    'clf__max_features': [0.5, 1],
    'clf__max_depth': [3, 4, 5],
}
gs = GridSearchCV(pipeline, params, cv=5, scoring='roc_auc', n_jobs=4)
gs.fit(rich_features, target)
```
